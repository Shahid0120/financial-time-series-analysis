```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

# Mean and covariance
mean1 = [0, 0]
mean2 = [3, 3]
cov = [[2, 1], [1, 2]]

# Create distributions
dist1 = multivariate_normal(mean=mean1, cov=cov)
dist2 = multivariate_normal(mean=mean2, cov=cov)

# Create grid and compute density
x, y = np.mgrid[-5:8:.01, -5:8:.01]
pos = np.dstack((x, y))

# Plot distributions
plt.figure(figsize=(8, 8))
plt.contour(x, y, dist1.pdf(pos), levels=5, colors='blue', alpha=0.5)
plt.contour(x, y, dist2.pdf(pos), levels=5, colors='red', alpha=0.5)
plt.title('Distributions with the Same Covariance Matrix')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal, t

# Mean and covariance
mean1 = [0, 0]
mean2 = [3, 3]
cov = [[2, 1], [1, 2]]

# Create distributions
dist1 = multivariate_normal(mean=mean1, cov=cov)
dist2 = multivariate_normal(mean=mean2, cov=cov)

# Create a t-distribution with the same covariance
dist3 = t(df=3, loc=mean1, scale=np.linalg.cholesky(cov))

# Create grid and compute density
x, y = np.mgrid[-5:8:.01, -5:8:.01]
pos = np.dstack((x, y))

# Plot distributions
plt.figure(figsize=(12, 6))

# Plot normal distributions
plt.subplot(1, 2, 1)
plt.contour(x, y, dist1.pdf(pos), levels=5, colors='blue', alpha=0.5, label='Normal Dist 1')
plt.contour(x, y, dist2.pdf(pos), levels=5, colors='red', alpha=0.5, label='Normal Dist 2')
plt.title('Normal Distributions with Same Covariance')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(['Normal Dist 1', 'Normal Dist 2'])

# Plot normal and t-distribution
plt.subplot(1, 2, 2)
plt.contour(x, y, dist1.pdf(pos), levels=5, colors='blue', alpha=0.5, label='Normal Dist 1')
plt.contour(x, y, dist3.pdf(pos), levels=5, colors='green', alpha=0.5, label='t-Dist')
plt.title('Normal and t-Distribution with Same Covariance')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(['Normal Dist 1', 't-Dist'])

plt.tight_layout()
plt.show()
```
