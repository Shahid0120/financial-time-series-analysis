```python
import numpy as np 
import scipy
import matplotlib.pyplot as plt
n = 10
d0 = 2 * np.ones(n)
d1 = -1 * np.ones(n-1)
A = np.diag(d0) + np.diag(d1, -1) + np.diag(d1, 1)
A
```

```python
sym_check = np.linalg.norm(A - A.T, 1)
sym_check
```

```python
plt.figure()
plt.spy(A)
plt.show()
```

```python
ev = scipy.linalg.eigvals(A)
ev = np.sort(ev)
x = np.arange(0, len(ev))
plt.figure()
plt.scatter(x, ev)
plt.show()
```

```python
evmax = ev.max()
evmin = ev.min()
cond = np.abs(evmax) / np.abs(evmin)
cond
```
