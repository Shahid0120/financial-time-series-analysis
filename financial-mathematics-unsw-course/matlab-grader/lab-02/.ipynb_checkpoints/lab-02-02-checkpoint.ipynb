```python
import numpy as np
import scipy 
import matplotlib.pyplot as plt
from numpy import empty, linspace, array, ones, log, exp
from scipy.linalg import lstsq, solve, qr, solve_triangular, svd, norm
from numpy.linalg import cond
from matplotlib.pyplot import figure, plot, xlim, xlabel, ylabel, grid, legend, title

data = np.loadtxt('lsdata.dat.txt', unpack=True)
np.shape(data)
```

```python
X = data[0,:].T
Y = data[1, :].T


```

```python
print(np.shape(X))
print(np.shape(Y))
```

```python
# A [ [1,...,1], [x,...,x], [x^2,.....,x^2], [x^3,....,x^3],[x^4,....,x^4]]
A = ones((1000,5))
A[:, 1] = X
A[:, 2] = X * X
A[:, 3] = X * X * X
A[:, 4] = X * X * X * X
np.shape(A)
```

```python
p, res, rnk, s = lstsq(A, Y)
p
# y = -2.198696 + 9.97581025x + -16.31362871x^2 + 10.65530806x^3 + 0.29997631x^4
```

```python
plt.figure()
plt.scatter(X,Y)
plt.ylabel('y values')
plt.xlabel('x values')
plt.show()
```

```python
x = np.arange(0, 1.2, 0.1)
ysol = -2.198696 + 9.97581025 * x + -16.31362871 * (x ** 2) + 10.65530806* (x ** 3) + 0.29997631* (x ** 4)
plt.figure()
plt.scatter(X,Y, s=2)
plt.plot(x, ysol, 'green')
plt.ylabel('y values')
plt.xlabel('x values')
plt.show()
```

```python
# for even x point is computes y and then ysol then adds to error 
ysol = -2.198696 + 9.97581025 * x + -16.31362871 * (x ** 2) + 10.65530806* (x ** 3) + 0.29997631* (x ** 4)
sysol = 0
approximation_err = 0 

ysol_arr = np.ones(1000)

for index, value in enumerate(X):
    x = value 
    ysol = -2.198696 + 9.97581025 * x + -16.31362871 * (x ** 2) + 10.65530806* (x ** 3) + 0.29997631* (x ** 4)
    approximation_err += abs(Y[index] * ysol)
    ysol_arr[index] = ysol
    sysol += ysol 



P = Y - ysol_arr
approx_diff = scipy.linalg.norm(P, 2)
approx_diff
```
