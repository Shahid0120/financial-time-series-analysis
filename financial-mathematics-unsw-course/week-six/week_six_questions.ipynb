```python
import numpy as np 
import scipy 
from numpy import loadtxt,  ones, empty, linspace, exp 
from numpy import zeros_like, pi, cos, sin, round, floor
from scipy.linalg import lstsq
from scipy.interpolate import InterpolatedUnivariateSpline, CubicSpline
from scipy.fft import fft
from matplotlib.pyplot import plot, figure, legend, grid, xlabel
import matplotlib.pyplot as plt

data = loadtxt('./gauss.dat.txt')
```

```python
plt.figure()
plt.scatter(data[:, 0], data[:,1])
plt.show()
print(data)
```

```python
# Least Sqaures with Basis functions y_hat = b0 * 1+ b1 * t + b2 * e^-(t-20^2/2) + b3 * -(t-7)^2/2
t = data[:,0]
Y = data[:,1]

# defining A
A = np.ones((len(t), 4))
A[:,1] = t
A[:, 2] = np.exp((-(t-3)**2)/2)
A[:,3] = np.exp((-(t-7)**2)/2)

b = scipy.linalg.lstsq(A, Y)
print(f"the coefficients are {b[0]}")
#
print(f"the residuals is {b[1]}")

Y_hat = []
x = 0
while x < 10.5:
    y_hat = 19.68872446 + 1.1114099 * x + 3.0196291* np.exp((-(x-3)**2)/2) + 4.88220957 * np.exp((-(x-7)**2)/2)
    Y_hat.append(y_hat)
    x += 0.5

plt.figure()
plt.scatter(t, Y)
plt.plot(t, Y_hat, color='red')
plt.title('Least Sqaures Approximartion using set basis functions')
plt.show()

```

```python
# Cublic Splines - not a knot end conditions
spl = CubicSpline(t,Y)
print(spl(t))
plt.figure()
plt.scatter(t, Y)
plt.plot(t, Y_hat, color='red')
plt.plot(t, spl(t), color='black')
plt.title('Least Sqaures Approximartion using set basis functions')
plt.show()
```

# Cublic Splines interpolation

```python
import scipy
from scipy.interpolate import CubicSpline

spl = CubicSpline(t, Y)
cspl = CubicSpline(t, Y, bc_type='clamped')
# plot
t_fine = np.linspace(min(t), max(t), 1000)
# Plot
plt.figure(figsize=(10, 6))
plt.plot(t_fine, spl(t_fine), color='red', label='S (Spline) not-a-knot')
plt.plot(t_fine, cspl(t_fine), color='green', label='S (Spline) clampped')
plt.plot(t, Y_hat, color='purple', label='lsqt')
plt.scatter(t, Y, color='blue', label='Data Points')
plt.legend()
plt.title('Cubic Spline Interpolation and Its Derivatives')
plt.xlabel('t')
plt.ylabel('Y / Derivatives')
plt.grid(True)
plt.show()
```
