# Question 4 part d
This question exames the our data fitting is affected by outliers 
suppose t_2 = 0.5 was corrupted into t_2 = 4.5

```python
import numpy as np 
import scipy
import matplotlib.pyplot as plt
from numpy import empty, linspace, array, ones, log, exp, reshape
from scipy.linalg import lstsq, solve, qr, solve_triangular, svd, norm
from numpy.linalg import cond
from matplotlib.pyplot import figure, plot, xlim, xlabel, ylabel, grid, legend, title

# data
tdat = np.array([0, 0.5, 1, 1.5, 2, 2.5, 3])
ydat = np.array([1.2, 1.5, 1.9, 2.4, 3.2, 4.3, 5.4])

tdat_corrupted = np.array([0, 4.5, 1, 1.5, 2, 2.5, 3])
ydat_corrupted = np.array([1.2, 1.5, 1.9, 2.4, 3.2, 4.3, 5.4])

# plot 
plt.figure() 
plt.scatter(tdat, ydat, color='blue', label='Original Data')
plt.scatter(tdat_corrupted, ydat_corrupted, color='green', label='Corrupted Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot of Original and Corrupted Data')
plt.legend()
plt.grid(True)
plt.show()

```
