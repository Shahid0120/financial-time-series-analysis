# Question 1(b) 
This question uses A = [1, t , t^2] still linear model

```python
import numpy as np 
import scipy
import matplotlib.pyplot as plt
from numpy import empty, linspace, array, ones, log, exp, reshape
from scipy.linalg import lstsq, solve, qr, solve_triangular, svd, norm
from numpy.linalg import cond
from matplotlib.pyplot import figure, plot, xlim, xlabel, ylabel, grid, legend, title

# defining variables
m = 7 
tdat = np.array([0, 0.5, 1, 1.5, 2, 2.5, 3])
ydat = np.array([1.2, 1.5, 1.9, 2.4, 3.2, 4.3, 5.4])

# A = [1, t, t^2]
AQ = np.ones((7,3))
AQ[:, 1] = tdat
AQ[:, 2] = tdat * tdat
print(AQ.shape)

# calculating Least Sqaures
xq = scipy.linalg.lstsq(AQ, ydat)
print(f"the coeffecients {xq[0]}")

# residuals
print(f"the residuals are {xq[1]}")
x = np.arange(0, 10, 0.1)
y_hat = []
for index in range(np.shape(ydat)[0]):
    approx = xq[0][0] + xq[0][1] * tdat[index] + xq[0][2] * (tdat[index] ** 2)
    y_hat.append(approx)

# plot
plt.figure()
plt.title('Figure 1')
plt.scatter(tdat, ydat)
plt.xlabel('tdat')
plt.ylabel('y')
plt.plot(tdat, y_hat, 'green', label='Quadratic Approximation')
plt.show()



```
