in this part we will approximate the data using exponential function z(a) = a*e^{miu * a}

# taking log both sides
$$
\begin{align*}
z(\lambda) &= \lambda e^{\mu \lambda} \leftrightarrow \\ 
log(z(\lambda)&= log(\lambda) + \mu \lambda
\end{align*}
$$
Now we have it in linear form !

```python
import numpy as np 
import scipy
import matplotlib.pyplot as plt
from numpy import empty, linspace, array, ones, log, exp, reshape
from scipy.linalg import lstsq, solve, qr, solve_triangular, svd, norm
from numpy.linalg import cond
from matplotlib.pyplot import figure, plot, xlim, xlabel, ylabel, grid, legend, title

# data
tdat = np.array([np.finfo(float).eps, 0.5, 1, 1.5, 2, 2.5, 3])
ydat = np.array([1.2, 1.5, 1.9, 2.4, 3.2, 4.3, 5.4])

# Feature Vector Initialisation 
A = np.ones((7,2))
A[:, 0] = np.log(tdat) 
A[:, 1] = tdat

# Solving Ax = y 
sol1 = np.linalg.lstsq(A, ydat)

# converting back to normal form 
param = [exp(sol1[0][0]), exp(sol1[0][1])] 


# approximating the y_hat using function above
def approx(x):
    param = [exp(sol1[0][0]), exp(sol1[0][1])] 
    return param[0] * exp(param[0] * x)
    
# array to store 
residual = []
y_hat_index = []

# loop to find y_hat and residual array 
for index, value in enumerate(ydat):
    y_hat = approx(tdat[index])
    y_hat_index.append(y_hat)
    residual.append(y_hat - value)

# calculate 2-norm of residuals
r1 = np.linalg.norm(residual)

print(f"the residual array for each index {residual} and \nthe 2-norm residual {r1}")

# plot 
plt.figure()
plt.scatter(tdat, ydat)
plt.plot(tdat, y_hat_index)
plt.show()
```
