```python
# question 2
import scipy 
import numpy as np 
import matplotlib.pyplot as plt 
f1 = lambda x: x**2
f2 = lambda x: np.exp(-x**2 / 4)

x = np.arange(0, 2)
plt.figure()
plt.plot(x, f1(x))
plt.plot(x, f2(x))
plt.show()
```

```python
# by hand the intersec is at x* = 0.8
f3 = lambda x: x**2 + 8*np.log(x)
x = np.arange(0.1, 3)
plt.figure()
plt.plot(x, f3(x))
plt.show()
```

```python
x0 = 0.8
devf3 = lambda x: 2*x + 8/x
x1 = x0 - (f3(x0)/ devf3(x0))
x1 = 0.899
print(f"|f1(x_1) - f_2(x_2)| = {np.abs(f1(x1) - f2(x1))}")

while np.abs(x1 - x0) > 2**(-52):
    x0 = x1
    x1 = x0 - (f3(x0)/ devf3(x0))

print(x1)
```
