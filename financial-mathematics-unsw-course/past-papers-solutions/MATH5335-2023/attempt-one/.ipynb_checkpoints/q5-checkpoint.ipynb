```python
import numpy as np 
import scipy
N = 4
h = 1/N
x = []
for index in range(0, N+1):
    x_n = index*h
    x.append(x_n)
print(x)
```

```python
# define 
row = 0
a = []
A = np.zeros((N,N))
A[row, 0] += 3
A[row, 1] += -4
A[row, 2] += 1 
a.append(0)
f = lambda x: (np.pi**2 /4 + 1) * np.cos(np.pi + np.pi*(x/2))
row = 1
for i in range(1, N):
    try:
        
        if i + 1 == 4:
            A[row, i] += -2 - h**2 
            A[row, i -1] += 1
        else:
            A[row, i] += -2 - h**2 
            A[row, i+1] += 1
            A[row, i -1] += 1
            
        a_val = h**2 * f(x[i])
        a.append(a_val)
        row += 1
    except IndexError as e:
        print(f"oh no {e}")
        print(A)

print(A)
print(a)
    
```

```python
norm = np.linalg.norm(A.T @ A - np.identity(N), 2)
norm
```

(2) Orthgonal -> since the norm > 0 clearly not orthogonal 
(3) LU factorisation !


```python
U4 = scipy.linalg.solve(A, a)
print(U4)
```
