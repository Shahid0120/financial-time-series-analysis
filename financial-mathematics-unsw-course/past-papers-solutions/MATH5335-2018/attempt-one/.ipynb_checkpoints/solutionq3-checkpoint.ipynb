```python
# Question 3
import numpy as np 
import scipy 
import matplotlib.pyplot as plt 
import blackscholes
from blackscholes import blackscholes

S = 100 
K = 90
r = 0.05
T = 1 
c = 30

g = lambda x: blackscholes(S,K,r,x,T)[0] - c
xd = np.arange(0, 20) / 20
xd = xd.T
print(xd)
yd = g(xd)
print(yd)
A = np.ones((len(xd), 6))
A[:, 1] = xd 
A[:, 2] = xd**2
A[:, 3] = xd**3
A[:, 4] = xd**4
A[:, 5] = xd**5

sol = scipy.linalg.lstsq(A, yd)
als = sol[0]
print(als)
cs = scipy.interpolate.CubicSpline(xd, yd)

xpl = np.linspace(0 ,1 , 101)
ysl = als[0] + als[1]*xpl + als[2]*(xpl**2) + als[3]*(xpl**3) + als[4]*(xpl**4) + als[5]*(xpl**5)
ysp = cs(xpl)

np.shape(ysp)

plt.figure()
plt.scatter(xd, yd)
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Cubic Spline vs Least Square Approximation')
plt.plot(xpl, ysl, label='Least Squares')
plt.plot(xpl, ysp, label='Cubic Spline')
plt.legend()
plt.show()
```
