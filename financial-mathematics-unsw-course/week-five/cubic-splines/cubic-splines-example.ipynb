# Adding Cubic interpolation

```python
import scipy
from scipy.interpolate import CubicSpline


# orginal graph
x = np.arange(-4, 4, 0.1)
f = 1 / (1 + x ** 2 )
    
# cubic interpolation
x = np.arange(-4, 4, 0.1)
f = 1 / (1 + x ** 2 )
cs = CubicSpline(x, f)
xs = np.arange(-4, 4, 0.5)

# plotting intervals for 0.5
x_points_intervals = np.arange(-4,4,0.5)
y_points_intervals = []
for index in range(len(x_points_intervals)):
    function_y = 1 / (1 + x_points_intervals[index] ** 2)
    y_points_intervals.append(function_y)
fig, ax = plt.subplots()

# plot original graph
ax.plot(x,f)


# plot end point of cubic splines
ax.scatter(x_points_intervals, y_points_intervals, color='red')

# plot cubic splines interpolation 
ax.plot(xs, cs(xs))

plt.show()

```
