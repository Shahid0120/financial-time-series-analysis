# Question 2 - Tri-diagonal Sparcity

```python
import numpy as np
# Create a triangular matrix
n = 10
e = np.ones(n)
A = 2*np.diag(e[:n]) + -1*np.diag(e[:n-1],-1) + -1*np.diag(e[:n-1],1)
A
```

```python
symchk = np.linalg.norm(A-A.T,1)
symchk
```

A is symetric since ||A - A.T||_1 = 0

```python
# find all A non-zero elements and graph them 
import scipy
import matplotlib.pyplot as plt
A_nonzero_elements = scipy.sparse.csr_matrix(A).data
index = np.arange(0, len(A_nonzero_elements))

# plot nonzero elements
unique_values_count = set(A_nonzero_elements)
plt.figure()
plt.spy(A,markersize=2)
plt.title("A spy plot")
plt.show()
# so it if is diagonal 2 = n, -1 = (n - 1)*2. which is true
```

```python
ev = np.sort(np.linalg.eigvals(A).real)
print(ev)
```

```python
x = np.arange(n)
plt.figure()
plt.scatter(x, ev)
plt.title('Eigenvalues for A')
plt.ylabel('Eigenvalues Magnitude')
plt.show()
```

```python
# max eignevalue
evmax = ev.max()
evmin = ev.min()
# since it is symetric matrix thus the condition number is k_2(a) = |eig max|/\eig min|
evcnd = abs(evmax) / abs(evmin)
```
