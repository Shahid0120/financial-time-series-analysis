```python
import matplotlib.pyplot as plt
import numpy as np

# Create an empty 36x36 grid
grid_size = 36
grid = np.zeros((grid_size, grid_size, 3), dtype=np.uint8)

# Define the box position and color
box_color = [255, 0, 0]  # Red color for the box
box_start = (0, 0)
box_end = (2, 2)

# Fill the box area with the specified color
for x in range(box_start[0], box_end[0] + 1):
    for y in range(box_start[1], box_end[1] + 1):
        grid[x, y] = box_color

# Display the grid with the box
fig, ax = plt.subplots(figsize=(6, 6))
ax.imshow(grid, interpolation='nearest')

# Set axis limits to range from 36 to -36
ax.set_xlim(0, grid_size - 1)
ax.set_ylim(grid_size - 1, 0)

# Set major ticks for the grid
ax.set_xticks(np.arange(0, grid_size, 1))
ax.set_yticks(np.arange(0, grid_size, 1))

# Set grid lines
ax.grid(which='both', color='gray', linestyle='-', linewidth=0.5)

# Invert the y-axis to range from 36 to -36
ax.invert_yaxis()

# Set the x-axis and y-axis labels
ax.set_xticklabels(np.arange(grid_size//2, -grid_size//2, -1))
ax.set_yticklabels(np.arange(grid_size//2, -grid_size//2, -1))

# Display the grid
plt.show()
```
